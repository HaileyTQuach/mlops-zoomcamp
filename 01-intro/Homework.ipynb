```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds()/60)
```

```python
filtered_df = df[(df.duration >= 1) & (df.duration <= 60)]
filtered_df.head()
```

```python
categorical = ['PULocationID', 'DOLocationID']
filtered_df[categorical].dtypes
```

```python
filtered_df.head()
```

```python
filtered_df[categorical] = filtered_df[categorical].astype(str)
filtered_df.head()
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=True)
```

```python
train_dicts = filtered_df[categorical].to_dict(orient='records')
```

```python
train_dicts
```

```python
X_train = dv.fit_transform(train_dicts)
```

```python
num_feats = X_train.shape[1]
num_feats
```

```python
target = 'duration'
y_train = filtered_df[target].values
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_train)
y_pred
```

```python
rmse = np.sqrt(mean_squared_error(y_pred,y_train))
rmse
```

```python
def read_dataframe(filename):
    if filename.endswith('.csv'):
        df = pd.read_csv(filename)

        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    elif filename.endswith('.parquet'):
        df = pd.read_parquet(filename)

    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    
    return df
```

```python
df_val = read_dataframe("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet")
```

```python
df_val.head()
```

```python
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID']
```

```python
val_dicts = df_val[categorical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
y_val = df_val[target].values
```

```python
y_pred = lr.predict(X_val)
```

```python
val_rmse = np.sqrt(mean_squared_error(y_pred,y_val))
val_rmse
```
